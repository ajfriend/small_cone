# Smallest cone enclosing a set of points

Given points $x_i \in \mathbf{R}^n$ for $n \geq 2$:

$$
\begin{array}{ll}
\text{minimize} & \log \det A^{-1}  \\
\text{subject to} & \|A x_i\|_2 \leq b^T x_i,\quad \forall i \\
& b^T A^{-1} b \leq 1
\end{array}
$$

# Feasibility

Above problem is feasible if

$$
\begin{array}{ll}
\text{minimize} & \| b \|_2  \\
\text{subject to} & 1 \leq b^T x_i, \forall i
\end{array}
$$

is feasible. (Don't actually need the objective, but it gives a nice separating hyperplane.)

# Related problem: Smallest enclosing circular cone

$$
\begin{array}{ll}
\text{minimize} & \| b \|_2  \\
\text{subject to} & \|x_i\|_2 \leq b^T x_i,\quad \forall i \\
\end{array}
$$

```python
def hex_cond(h):
    x = np.array([
        ll23d(math.pi/2 - lat*math.pi/180, lng*math.pi/180)
        for lat, lng in 
        h3.h3_to_geo_boundary(h)
    ])

    A, b, prob = u.small_cone(x)
    q = u.perp(b)
    return np.linalg.cond(q@A@(q.T))

import h3
import math

def ll23d(lat, lng):
    x = math.sin(lat) * math.cos(lng)
    y = math.cos(lat)
    z = math.sin(lat) * math.sin(lng)
    
    return x, y, z


hexes = h3.get_res0_indexes()
# hexes = set.union(*[
#     h3.h3_to_children(h)
#     for h in hexes
# ])


out = {}
for h in hexes:
    out[h] = hex_cond(h)
```
