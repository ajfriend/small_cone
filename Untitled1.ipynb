```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

import pandas as pd
import cvxpy as cvx
```

```python
def get_demand():
    N = 100

    vals = [
        (0, 0),
        (20, .5),
        (50, -.5),
        (60, -.1),
        (70, -.2),
        (80, .0),
        (85, 0),
        (87, 0),
    ]


    y = cvx.Variable(N)
    x = np.linspace(0,1, N)

    obj = 0
    for i, v in vals:
        obj = obj + cvx.square(y[i] - v)

    obj = obj + .2*cvx.norm(cvx.diff(y, 3))

    prob = cvx.Problem(cvx.Minimize(obj))
    prob.solve()
    y = y.value
    
    return x, y

def get_supply():
    N = 100

    vals = [
        (0, 0),
        (15, 0),
        (25, 0),
        
        (30, .3),
        (35, -.3),
        (37, -.3),
        (39, -.3),
        
        (70, -.2),
        (80, -.2),
        (85, 0),
        (87, 0),
    ]

    y = cvx.Variable(N)
    x = np.linspace(0,1, N)

    obj = 0
    for i, v in vals:
        obj = obj + cvx.square(y[i] - v)

    obj = obj + .2*cvx.norm(cvx.diff(y, 3))

    prob = cvx.Problem(cvx.Minimize(obj))
    prob.solve()
    y = y.value
    
    return x, y


def get_etd():
    N = 100

    vals = [
        (0, 0),
        (15, 0),
        (25, 0),
        
        (30, -.3),
        (35, .3),
        (37, .3),
        (39, .3),
        
        (70, .2),
        (80, .2),
        (85, 0),
        (87, 0),
    ]

    y = cvx.Variable(N)
    x = np.linspace(0,1, N)

    obj = 0
    for i, v in vals:
        obj = obj + cvx.square(y[i] - v)

    obj = obj + .2*cvx.norm(cvx.diff(y, 3))
    cst = [y >= 0, y <= .5]

    prob = cvx.Problem(cvx.Minimize(obj), cst)
    prob.solve()
    y = y.value
    
    return x, y
```

```python
x, y_demand = get_demand()
_, y_supply = get_supply()
_, y_etd = get_etd()
```

```python
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

fig, ax = plt.subplots(3, 1, figsize=(7,5), sharex=True, sharey=False)

f = np.log(x + 1)*.5

ax[0].plot(x, y_demand, label='demand')
ax[0].fill_between(x, y_demand - f, y_demand + f, alpha=.3, label='uncertainty', color=colors[0])
ax[0].legend()
ax[0].get_yaxis().set_visible(False)
ax[0].get_xaxis().set_visible(False)

ax[1].plot(x, y_supply, label='supply', color = colors[1])
ax[1].fill_between(x, y_supply - f, y_supply + f, alpha=.3, label='uncertainty', color = colors[1] )
ax[1].legend()
ax[1].get_yaxis().set_visible(False)
ax[1].get_xaxis().set_visible(False)


ax[2].plot(x, y_etd, label='ATD', color = colors[3])
ax[2].fill_between(x, y_etd - f, y_etd + f, alpha=.3, label='uncertainty', color = colors[3])
ax[2].legend()
ax[2].get_yaxis().set_visible(False)
ax[2].get_xaxis().set_visible(False)

#fig.savefig('out.png', dpi=600)
```
